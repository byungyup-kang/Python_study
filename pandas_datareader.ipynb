```python
from pandas_datareader import data as web
import pandas as pd
import numpy as np

start_date = '2010-01-01'
company_arr = ['035420.KS', '067160.KQ'] 
df1 = web.DataReader(company_arr[0], 'yahoo', start_date)
df2 = web.DataReader(company_arr[1], 'yahoo', start_date)
```

```python
df1.head(20)
```

```python
open_weekly = df1['Open'].resample('W').first()
close_weekly = df1['Close'].resample('W').last()
high_weekly = df1['High'].resample('W').max()
low_weekly = df1['Low'].resample('W').min()
price_weekly = df1['Adj Close'].resample('W').mean()
df_weekly = pd.concat([open_weekly, high_weekly, low_weekly, close_weekly, price_weekly], axis=1)
```

```python
df_weekly.head()
```

```python
price_ewm = df_weekly['Adj Close'].ewm(com=20)
ewm_mean = price_ewm.mean()
ewm_std = price_ewm.std()
```

```python
tmp = np.sqrt(10)*(df_weekly['Adj Close'] - ewm_mean)/ewm_std
```
